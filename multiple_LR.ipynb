
**Hypothesis function using matrix multiplication**


   where 

**Cost Function for multivriate Linear regression is:**


```python
def J(Y,Hx):
    return np.sum(np.square(Y-Hx)) / 2 / Y.shape[0]
```

**gradient descent**


```python
X = np.array(adds[['TV','Radio','Newspaper']])
Y = np.array(adds['Sales'])[:,np.newaxis]
```

```python
def gradient_descent(X,Y,n,starting_thetas,alpha,iterations):
    m = Y.shape[0]
    if starting_thetas is None:
        starting_thetas = [0] * (n+1)
    thetas = np.array(starting_thetas)[:,np.newaxis]
    
    ones = np.ones(m)[:,np.newaxis]
    #Add ones in front of X for theta_0
    X2 = np.concatenate((ones,X),axis=1)
    costs = []
    for i in range(0,iterations):
        Hx = np.dot(X2,thetas)
        cost = J(Y,Hx)
        dy = -1/Y.shape[0] * np.dot(X2.T,Y-Hx)
        thetas = thetas - alpha * dy

        costs.append(cost)
        
        
    return thetas,costs
```

```python
# Set hyperparameters
alpha = 0.000001
iterations = 200
n = X.shape[1]
starting_thetas = None
```

```python
thetas, costs = gradient_descent(X,Y,n,starting_thetas,alpha,iterations)
print("Thetas: {0}".format(thetas))
#print(costs)
plt.plot(costs)
plt.ylabel('cost')
plt.xlabel('iteration')
plt.show()
```

```python
from mpl_toolkits import mplot3d
from matplotlib import cm
import matplotlib.pyplot as plt

ones = np.ones(200)[:,np.newaxis]
X2 = np.concatenate((ones,X),axis=1)

fig = plt.figure(figsize=(14,7))
ax  = plt.axes(projection='3d')

X0,X1 = np.meshgrid(X2.T[1],X2.T[2])
yval = thetas[0] + X0 * thetas[1] + X1 * thetas[2]

ax.plot_surface(X0,X1,yval,cmap=cm.coolwarm,alpha=0.5)

ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')
plt.title('Visualization of predicted sales values on 2 features')

plt.show()
```
